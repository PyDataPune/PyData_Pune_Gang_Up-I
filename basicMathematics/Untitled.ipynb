```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-2, 2,0.01)
plt.plot(x, x, linewidth=3.0, c='r')
plt.axhline(y=0.0, linewidth=2, c='k')
plt.axvline(x=0.0, linewidth=2, c='k')
plt.grid()
plt.savefig('fig1.png')
```

```python
plt.plot(x, x, linewidth=3.0, c='r')
plt.plot(x, x**2, linewidth=3.0, c='royalblue')
plt.axhline(y=0.0, linewidth=2, c='k')
plt.axvline(x=0.0, linewidth=2, c='k')
plt.grid()
plt.savefig('fig2.png')
```

```python
plt.plot(x, x, linewidth=3.0, c='r')
plt.plot(x, x**2, linewidth=3.0, c='royalblue')
plt.plot(x, x**3, linewidth=3.0, c='orange')
plt.axhline(y=0.0, linewidth=2, c='k')
plt.axvline(x=0.0, linewidth=2, c='k')
plt.grid()
plt.savefig('fig3.png')
```

```python
plt.plot(x, np.exp(x), linewidth=3.0, c='royalblue')
plt.axhline(y=0.0, linewidth=2, c='k')
plt.axvline(x=0.0, linewidth=2, c='k')
plt.grid()
plt.savefig('fig4.png')
```

```python
z = np.arange(0.01,10,0.01)
plt.plot(z, np.log(z), linewidth=3.0, c='royalblue')
plt.axhline(y=0.0, linewidth=2, c='k')
plt.axvline(x=0.0, linewidth=2, c='k')
plt.grid()
plt.savefig('fig5.png')
```
